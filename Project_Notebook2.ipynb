My first question ask how does the implementation of women-run WHD impacts the number of First Information Reports(FIR) filed by women. Here I explore whether if the amount of reports increase or whether they decrease.

```python
#Reviewing our variable "treatment" to see how many police stations implemented a woman-run WHD.
Police_full_df['treatment'].value_counts()
```

##Univariate Analysis

```python
## Plotting the bar chart
Police_full_df['treatment'].value_counts().plot(
    kind='bar',
    color=['skyblue', 'salmon'],
    figsize=(8, 5)
)

# Adding titles and labels
plt.title('Treatment Distribution', fontsize=16)
plt.xlabel('Treatment Decision', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

This shows us that there are twice as many police stations that have implemented women-run help desks. This is a positive, although there is a large number of stations without women-run WHD.

```python
# Plotting the bar chart for FIR by women
Admin_long_df['fir_bywomen_count'].value_counts().sort_index().plot(
    kind='bar',
    color='lightcoral',
    figsize=(8, 5)
)

# Adding titles and labels
plt.title('Distribution of FIRs Filed by Women', fontsize=16)
plt.xlabel('FIR Count by Women', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

The frequency of reports decreases as the amount of reports increases. The frequency goes from as often as 800 times for one report compared to locations that had eight to eleven reports.

```python
# Creating a line plot to show the difference in FIR reporting by month.
Admin_long_df.groupby('month')['fir_bywomen_count'].mean().plot(
    kind='line',
    marker='o',
    color='skyblue',
    figsize=(10, 6)
)

# Adding titles and labels
plt.title('FIRs Filed by Women by Month', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Average FIRs Filed by Women', fontsize=12)
```

The average amount of reports increased to its highest point during the spring and summer, but drastically decreased after January 2020. I would assume the reason for the drastic change was the pandemic, and reports or police interactions were not the same and reports were filed less often.


```python
#Total amount of fir_bywomen_count
Admin_long_df['fir_bywomen_count'].sum()
```

It's clear from the graph that 62% believe the police are very helpful, followed by 36% showing the same sentiments. There is not even 1% from the survey who think the police are unhelpful.

## Quick data clean and merging of dataframes

```python
#Polishing dataframes for further use
Police_full_df = Police_full_df.loc[:, ['b_effective', 'e_effective', 'treatment']]

Police_base_df = Police_base_df.loc[:, ['b_effective', 'b_helpful', 'treatment']]

Citizen_full_df = Citizen_full_df.loc[:, ['b_visit', 'e_visit', 'b_pol_handling', 'e_pol_handling', 'member_gender', 'treatment']]
```

```python
#Merging Police Full and Police_base DataFrames
police_df = pd.merge(Police_full_df, Police_base_df, on='treatment', how='inner')
```

```python
#Checking to confirm merger worker
police_df.head()
```

```python
#Checking Citizen dataframe to ensure the clean up worked
Citizen_full_df.head()
```

```python
#Dropped duplicates from Police dataframe
police_df.drop_duplicates(inplace=True)
```

```python
# Calculating the mean of the 'e_visit' column
mean_e_visit = Citizen_full_df['e_visit'].mean()

# Calculating the mean of the 'e_pol_handling' column
mean_e_pol_handling = Citizen_full_df['e_pol_handling'].mean()

#  these calculated means to fill NaN values
Citizen_full_df.fillna(value={'treatment': 0, 'e_visit': mean_e_visit, 'e_pol_handling': mean_e_pol_handling}, inplace=True)
```

```python
#Checking to confirm values were filled correctly
Citizen_full_df.head()
```
